```python
class BaseModel :
    @staticmethod
    def process(obj):
        
        return obj.step1() + obj.step2()
        
    def step1(self):
        
        return "base1"
    def step2(self):
        
        return "base2"
        
class CustomModel(BaseModel):
    
    def step2(self):
        return "custom2"   
        
m = CustomModel()

print(m.process(m))
```
